```python
#Import Packages and dataframe:

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (12,8)

ds = pd.read_csv(r'c:\M395_weekly_sales .csv')
```

```python
ds.head()
```

```python
# Checking for Missing Values in Data and dropping Null Values:

for col in ds.columns:
    p_missing = np.mean(ds[col].isnull())
    print('{} = {}%'.format(col, p_missing))

ds = ds.dropna()
```

```python
#Analyzing the data types for each column:

ds.dtypes
```

```python
#Data visualization: Scatter Plot 

plt.scatter(x = ds['REST_KEY'], 
            y = ds['adus'])
plt.xlabel('Restaurant Key')
plt.ylabel('Average Daily Units Sold')
plt.title('Effect of Restaurant Key on Daily Units Sold')

plt.show
```

```python
#Regression Analysis using Seaborn package:

sns.regplot(x='REST_KEY',y ='adus',data=ds, scatter_kws={"color": "Purple"}, line_kws={"color": "Blue"}) 
plt.xlabel('Restaurant Key')
plt.ylabel('Average Daily Units Sold')
plt.title('Effect of Restaurant Key on Daily Units Sold')

```

```python
df = pd.read_csv(r'c:\M395_rest_facts.csv')
```

```python
df.head()
```

```python
# Checking for Missing Values in Data and dropping Null Values in Restaurant Facts:

for col in df.columns:
    p_missing = np.mean(df[col].isnull())
    print('{} = {}%'.format(col, p_missing))

df = df.dropna()
```

```python
#Analyzing the data types for each column:

df.dtypes
```

```python
#Creating Value Counts For Bar Chart Distribution of Restaurant Demographic Facts:

incomeq_label_counts = df['incomeq_label'].value_counts()
urban_label_counts = df['urban_label'].value_counts()
social_label_counts = df['social_label'].value_counts()
```

```python
#Plotting Income Level, Urban Label, and Social Label on a Bar Chart to Visualize Current Distribution of Restaurants:

plt.figure(figsize=(12, 6))
plt.subplot(131)
sns.barplot(x=incomeq_label_counts.index, y=incomeq_label_counts.values)
plt.title('Income Level Distribution')
plt.xticks(rotation=45)

plt.subplot(132)
sns.barplot(x=urban_label_counts.index, y=urban_label_counts.values)
plt.title('Distribution of Restaurants in Differing Urban Environments')
plt.xticks(rotation=45)

plt.subplot(133)
sns.barplot(x=social_label_counts.index, y=social_label_counts.values)
plt.title('Distribution of Restaurants in differing Social Labels')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
itemdesc_counts = ds['itemdesc'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
top_10_items = itemdesc_counts[:10]
sns.barplot(x=top_10_items.index, y=top_10_items.values)
plt.title('Top 10 Selling Items')
plt.xticks(rotation=90)
plt.ylim(0, top_10_items.max() * 1.2) 

plt.tight_layout()
plt.show()
```

```python
#Merging the Dataframes together on a primary key column for further analysis:
merged_df = pd.merge(df, ds, on='REST_KEY')
merged_df.head()
```

```python
#Pivot table/Heatmap Creation:

pivot_table = pd.crosstab(merged_df['social_label'], merged_df['itemdesc'])

plt.figure(figsize=(16, 12))
sns.set(font_scale=0.5)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', cbar=True, linewidths=0.5)
plt.title('Social Label vs. Item Purchases')
plt.xticks(rotation=45)
plt.show()
```
